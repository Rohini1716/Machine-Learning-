# **üöó Introduction**

The dataset contains detailed information about used cars listed in Bangladesh, providing insights into the country‚Äôs second-hand automobile market. It consists of 1,500 records and 12 attributes, covering key features such as car brand and model, year of manufacture, mileage, engine capacity, ownership history, location, and both new and resale prices.

This dataset is valuable for:

Market Analysis ‚Üí Understanding pricing trends and demand in different cities.

Predictive Modeling ‚Üí Building machine learning models to predict resale car prices.

Consumer Insights ‚Üí Identifying how factors like fuel type, transmission, mileage, and ownership influence resale value.

Business Applications ‚Üí Assisting car dealers, buyers, and sellers in making data-driven decisions.

# **üìâ Linear Regression**

```python
y=data['price']
x=data.drop(['price'],axis=1)
```

```python
x.columns = x.columns.map(str)
```

```python
from sklearn.model_selection import train_test_split
x_test, x_train, y_test, y_train=train_test_split(x,y,train_size=0.8, random_state=42)
x_train.shape,x_test.shape,y_test.shape,y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, mean_absolute_error, r2_score
```

```python
print("Linear Regression Model Results")
print("Mean_squared_error value:", mean_squared_error(y_test,y_pred))
print("Mean_Absolute_percentage error value:", mean_absolute_percentage_error(y_test,y_pred))
print("Mean_Absolute_error value:", mean_absolute_error(y_test,y_pred))
print("R¬≤ Score value:", r2_score(y_test,y_pred))
```
